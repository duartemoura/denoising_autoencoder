## Define Autoencoder

In this section we are defining a 3 and 5 layer autoencoder to experiment with different paramteres. The goal is trying to understand which of these autoencoders works better for the dataset first, and the use the best one to denoise the imeages.

### 3 layer

```python
class Autoencoder3(nn.Module):

    def __init__(self, encoded_space_dim, h1, h2):
        super().__init__()

        ### Convolutional section
        self.encoder = torch.nn.Sequential(
            torch.nn.Linear(28 * 28, h1),
            torch.nn.ReLU(),
            torch.nn.Linear(h1, h2),
            torch.nn.ReLU(),
            torch.nn.Linear(h2, encoded_space_dim)
        )

        self.decoder = torch.nn.Sequential(
            torch.nn.Linear(encoded_space_dim, h2),
            torch.nn.ReLU(),
            torch.nn.Linear(h2, h1),
            torch.nn.ReLU(),
            torch.nn.Linear(h1, 28 * 28),
            torch.nn.Sigmoid(),
        )


    def forward(self, x):
        # Flatten the image from [batch_size, 1, 28, 28] to [batch_size, 28*28]
        x = x.view(x.size(0), -1)
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        output = decoded.view(x.size(0), 1, 28, 28)
        return output, encoded
```

### 5 layer

```python
import torch
import torch.nn as nn

class Autoencoder5(nn.Module):
    def __init__(self, encoded_space_dim, h1, h2, h3, h4):
        super().__init__()

        ### Encoder section
        self.encoder = nn.Sequential(
            nn.Linear(28 * 28, h1),
            nn.ReLU(),
            nn.Linear(h1, h2),
            nn.ReLU(),
            nn.Linear(h2, h3),
            nn.ReLU(),
            nn.Linear(h3, h4),
            nn.ReLU(),
            nn.Linear(h4, encoded_space_dim)
        )

        ### Decoder section
        self.decoder = nn.Sequential(
            nn.Linear(encoded_space_dim, h4),
            nn.ReLU(),
            nn.Linear(h4, h3),
            nn.ReLU(),
            nn.Linear(h3, h2),
            nn.ReLU(),
            nn.Linear(h2, h1),
            nn.ReLU(),
            nn.Linear(h1, 28 * 28),
            nn.Sigmoid()  # For pixel values between 0 and 1
        )

    def forward(self, x):
        # Flatten the image from [batch_size, 1, 28, 28] to [batch_size, 28*28]
        x = x.view(x.size(0), -1)
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        # Reshape the output back to [batch_size, 1, 28, 28]
        output = decoded.view(x.size(0), 1, 28, 28)
        return output, encoded
```

```python
def train_epoch(model, train_loader, criterion, optimizer, device, apply_noise=False, variance=None, lambda_l1=1e-5):
    model.train()
    running_loss = 0.0

    for data in tqdm(train_loader, leave=False, desc="Training"):
        inputs, _ = data  # Ignore labels
        inputs = inputs.to(device)
        optimizer.zero_grad()

        # Add noise to inputs if apply_noise is True
        if apply_noise:
            noisy_inputs = add_noise(inputs, variance)
        else:
            noisy_inputs = inputs

        # Forward pass with (potentially) noisy inputs
        outputs, encoded = model(noisy_inputs)

        # Compute loss between outputs and clean inputs
        loss_reconstruction = criterion(outputs, inputs)

        # L1 regularization on encoded representations
        loss_l1 = torch.mean(torch.abs(encoded))

        # Total loss
        loss = loss_reconstruction + lambda_l1 * loss_l1

        # Backward pass and optimization
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    avg_loss = running_loss / len(train_loader)
    return avg_loss
```

```python
def evaluate_model(model, data_loader, criterion, device, apply_noise=False, variance=None):
    model.eval()
    running_loss = 0.0
    total_psnr = 0.0

    with torch.no_grad():
        for data in tqdm(data_loader, leave=False, desc="Evaluating"):
            inputs, _ = data
            inputs = inputs.to(device)

            # Optionally add noise
            if apply_noise:
                noisy_inputs = add_noise(inputs, variance)
            else:
                noisy_inputs = inputs

            outputs, _ = model(noisy_inputs)

            # Clamp outputs to valid range
            outputs = outputs.clamp(0, 1)

            # Compute loss
            mse_loss = criterion(outputs, inputs)
            running_loss += mse_loss.item()

            # Compute PSNR
            if mse_loss.item() == 0:
                psnr = float('inf')
            else:
                psnr = 20 * torch.log10(1.0 / torch.sqrt(mse_loss))
            total_psnr += psnr.item()

    avg_loss = running_loss / len(data_loader)
    avg_psnr = total_psnr / len(data_loader)
    return avg_loss, avg_psnr
```

```python
def train_autoencoder(model, train_loader, test_loader, num_epochs, learning_rate, lambda_l1, device, apply_noise=False, variance=None):
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    train_losses = []
    test_losses = []
    test_psnr_list = []

    model.to(device)

    for epoch in range(num_epochs):
        # Train for one epoch
        avg_train_loss = train_epoch(
            model, train_loader, criterion, optimizer, device,
            apply_noise=apply_noise, variance=variance, lambda_l1=lambda_l1
        )
        train_losses.append(avg_train_loss)

        # Evaluate on test set
        avg_test_loss, avg_psnr = evaluate_model(
            model, test_loader, criterion, device,
            apply_noise=apply_noise, variance=variance
        )
        test_losses.append(avg_test_loss)
        test_psnr_list.append(avg_psnr)

        # Log every epoch to the log file
        logging.info(f"Epoch [{epoch+1}/{num_epochs}], "
                     f"Train Loss: {avg_train_loss:.4f}, "
                     f"Test Loss: {avg_test_loss:.4f}, "
                     f"Test PSNR: {avg_psnr:.2f} dB")

        # Print only the first and last epochs to the console
        if epoch == 0 or epoch == num_epochs - 1:
            print(f"Epoch [{epoch+1}/{num_epochs}], "
                  f"Train Loss: {avg_train_loss:.4f}, "
                  f"Test Loss: {avg_test_loss:.4f}, "
                  f"Test PSNR: {avg_psnr:.2f} dB")

    return train_losses, test_losses, test_psnr_list
```

```python
def visualize_reconstructions(model, data_loader, device, title, dataset_name, apply_noise=False, variance=None):
    model.eval()
    with torch.no_grad():
        dataiter = iter(data_loader)
        images, _ = next(dataiter)
        images = images.to(device)

        # Add noise to images if required
        if apply_noise:
            noisy_images = add_noise(images, variance)
            inputs = noisy_images
        else:
            inputs = images

        outputs, _ = model(inputs)

        # Move data to CPU for plotting
        images = images.cpu()
        outputs = outputs.cpu()
        if apply_noise:
            noisy_images = noisy_images.cpu()

        # Determine the number of rows based on whether noise is applied
        num_rows = 3 if apply_noise else 2

        # Plot original, (noisy), and reconstructed images
        num_images = 6
        fig, axes = plt.subplots(num_rows, num_images, figsize=(num_images * 1.5, num_rows * 1.5))

        for i in range(num_images):
            # Original images
            axes[0, i].imshow(images[i].squeeze(), cmap='gray')
            axes[0, i].axis('off')

            if apply_noise:
                # Noisy images
                axes[1, i].imshow(noisy_images[i].squeeze(), cmap='gray')
                axes[1, i].axis('off')

                # Reconstructed images
                axes[2, i].imshow(outputs[i].squeeze(), cmap='gray')
                axes[2, i].axis('off')
            else:
                # Reconstructed images when no noise is applied
                axes[1, i].imshow(outputs[i].squeeze(), cmap='gray')
                axes[1, i].axis('off')

        # Add titles to the first column only
        if apply_noise:
            axes[0, 0].set_ylabel('Original', fontsize=12)
            axes[1, 0].set_ylabel('Noisy', fontsize=12)
            axes[2, 0].set_ylabel('Reconstructed', fontsize=12)
        else:
            axes[0, 0].set_ylabel('Original', fontsize=12)
            axes[1, 0].set_ylabel('Reconstructed', fontsize=12)

        # Add a single title for the entire figure
        plt.suptitle(title, fontsize=16)
        plt.tight_layout()
        plt.subplots_adjust(top=0.88)  # Adjust to make room for the suptitle

        # Save the plot
        # Format variance for filename (e.g., 0.01 -> '0p01')
        model_name = type(model).__name__
        variance_str = f"{variance}".replace('.', 'p') if variance is not None else 'no_noise'
        plot_filename = f'dataset_{dataset_name}_model_{model_name}_var_{variance_str}.png'
        plt.savefig(plot_filename)
        plt.show()
        plt.close()
```

```python
# List to store results
epoch_results = []

# Define architectures
architectures = ['Autoencoder3', 'Autoencoder5']

encoded_space_dims = [15, 30, 50, 100]

batch=128

dataloaders = createDataloaders(batch_size=batch, validation_split=0.2)

# Training parameters
num_epochs = 25
learning_rate = 1e-3
lambda_l1 = [1e-5, 1e-4, 1e-3]
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

for dataset_name, architecture, encoded_space_dim, l in product(dataloaders.keys(), architectures, encoded_space_dims, lambda_l1):
    print(f"\nProcessing dataset: {dataset_name}")

    # Retrieve the preloaded data loaders
    train_loader = dataloaders[dataset_name]['train_loader']
    test_loader = dataloaders[dataset_name]['test_loader']

    print(f"\nTraining {architecture} with encoded space dimension {encoded_space_dim} and L1 regularization of {l}")

    # Initialize the model
    if architecture == 'Autoencoder3':
        h1, h2 = 512, 256
        model = Autoencoder3(encoded_space_dim, h1, h2)
    elif architecture == 'Autoencoder5':
        h1, h2, h3, h4 = 656, 512, 256, 128 #so the dimmensions are never bigger than the encoded dimmension
        model = Autoencoder5(encoded_space_dim, h1, h2, h3, h4)
    else:
        continue  # Skip unknown architectures


    # Train the model
    train_losses, test_losses, test_psnr_list = train_autoencoder(
        model,
        train_loader,
        test_loader,
        num_epochs,
        learning_rate,
        l,
        device,
        apply_noise=False  # No noise for this comparison
    )

    # Store per-epoch results
    for epoch in range(num_epochs):
        epoch_results.append({
            'Dataset': dataset_name,
            'Architecture': architecture,
            'Encoded_Dim': encoded_space_dim,
            'L1_Penalty': l,
            'Epoch': epoch + 1,
            'Train_Loss': train_losses[epoch],
            'Test_Loss': test_losses[epoch],
            'Test_PSNR': test_psnr_list[epoch]
        })

# Convert results to DataFrame
epoch_results_df = pd.DataFrame(epoch_results)
print(epoch_results_df.head())
```

### Best architecture

From the plots we got from these experiments we can see that for the MNIST and FNMNIST dataset the best model is always the 3 layer autoencoder with encoded space dimmension = 100

#### Define input noise function (based on variance)

```python
def add_noise(inputs, variance=0.01):
    std_dev = torch.sqrt(torch.tensor(variance))
    noise = inputs + torch.randn_like(inputs) * std_dev
    noise = torch.clamp(noise, 0., 1.)
    return noise
```

### Denoising

Finally, we use the best model found previously to denoise the images. Here we can just pass the best model and now set the parameters of **apply_noise** to *True* and evaluating the model based on the variance

```python
num_epochs = 15

variances = [0.01, 0.02, 0.05, 0.1]

l_best = 0.0001

epoch_results_dn = []

for dataset_name in dataloaders.keys():
    print(f"\nProcessing noisy dataset: {dataset_name}")

    # Retrieve the preloaded data loaders
    train_loader = dataloaders[dataset_name]['train_loader']
    test_loader = dataloaders[dataset_name]['test_loader']

    for variance in variances:
        print(f"\nTraining with noise variance: {variance}")

        #Initialize model here to reset weights
        best_model = Autoencoder3(100, 512, 256) # The best model was the same in both datasets


        train_losses, test_losses, test_psnr_list = train_autoencoder(
            best_model,
            train_loader,
            test_loader,
            num_epochs,
            learning_rate,
            l_best,
            device,
            apply_noise=True,
            variance=variance
        )

        # Visualize the results
        visualize_reconstructions(
            best_model,
            test_loader,
            device,
            title=f"Denoising Autoencoder ({dataset_name}, Variance={variance})",
            apply_noise=True,
            variance=variance,
            dataset_name = dataset_name
        )

        # Store per-epoch results
        for epoch in range(num_epochs):
            epoch_results_dn.append({
                'Dataset': dataset_name,
                'Epoch': epoch + 1,
                'Variance': variance,
                'Train_Loss': train_losses[epoch],
                'Test_Loss': test_losses[epoch],
                'Test_PSNR': test_psnr_list[epoch]
            })

# Convert results to DataFrame
epoch_results_dn_df = pd.DataFrame(epoch_results_dn)
print(epoch_results_dn_df.head())
```
